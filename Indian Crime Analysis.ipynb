5 different types of datasets are provided for a period of year namely: Abduction, Accidents, Crimes, Murder, Rape

## analysing abduction data

```python
sum_list=[]
for i in range(len(abduction_data)):
    sum_list.append(abduction_data.iloc[i,2:].mean())
abduction_data["Total"]=sum_list
abduction_data.head()
```

### Abduction cases comparison

#### States and total number of abduction cases there

```python
plt.figure(figsize=(15,5))
chart=sns.barplot(abduction_data["STATE/UT"],abduction_data["Total"])
chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
chart
```

#### relation with offender

```python
for_pie=[]
for col in abduction_data.columns[2:-1]:
    for_pie.append(abduction_data[col].mean())
```

```python
patches, texts, autotexts=plt.pie(x=for_pie,labels=abduction_data.columns[2:-1],radius=2,autopct='%1.1f%%')
for i in range(len(autotexts)):
    autotexts[i].set_fontsize(10)
    texts[i].set_fontsize(10)
plt.show()
```

### Analysis:
#### (Through bar chart)<br>
The maximum number of abduction cases are reported in Madhya Pradesh.<br>
The number of abduction cases in Maharashtra, Rajasthan and Uttar Pradesh are also significantly high.<br>

#### (Through pie chart)<br>
In 50% of the cases reported, offenders were known to the victims.<br>
In around 25% of all cases, offender were other known persons to the victims

## Analysising accident data

```python
sum_list=[]
for i in range(1,len(accidents_data.columns)):
    sum_list.append(accidents_data.iloc[:,i].sum())
    
acc_mod_data=pd.DataFrame(data=[sum_list],columns=accidents_data.columns[1:])
acc_mod_data
```

```python
acc_plot_data=pd.DataFrame(index=["Fatal Accidents","Total Accidents","Killed","Injured"])
acc_plot_data["2012"]=[acc_mod_data.iloc[:,0][0],acc_mod_data.iloc[:,1][0],acc_mod_data.iloc[:,2][0],acc_mod_data.iloc[:,3][0]]
acc_plot_data["2013"]=[acc_mod_data.iloc[:,4][0],acc_mod_data.iloc[:,5][0],acc_mod_data.iloc[:,6][0],acc_mod_data.iloc[:,7][0]]
acc_plot_data["2014"]=[acc_mod_data.iloc[:,8][0],acc_mod_data.iloc[:,9][0],acc_mod_data.iloc[:,10][0],acc_mod_data.iloc[:,11][0]]
acc_plot_data["2015"]=[acc_mod_data.iloc[:,12][0],acc_mod_data.iloc[:,13][0],acc_mod_data.iloc[:,14][0],acc_mod_data.iloc[:,15][0]]

acc_plot_data
```

### Analysis:
The no. of crimes against women increased sharply in the span of 10 years from (2001-12).<br>

Uttar Pradesh, Andhra Pradesh and Maharashtra are major hotspots for crimes against women.<br>
